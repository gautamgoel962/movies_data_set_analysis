```python
# making a function to check the percent of null values in each column in a dataframe 
def check_percent_null_values(df):
    for col in df.columns:
        pct_missing = np.mean(df[col].isnull())
    #     print('{} - {}%'.format(col, pct_missing))
        print(f"{col}-{pct_missing}%")
```

```python
check_percent_null_values(df)
```

```python
# We, can see that null values exists 
# dropping all the rows which contain null values 
df = df.dropna()
```

```python
df.shape
```

```python
# Now, all the rows with at least one null value are dropped 
check_percent_null_values(df)
```

```python
sns.heatmap(correlation_matrix, annot = True, linewidths = 1, linecolor = "yellow")
plt.title("Correlation Matrix for numberic features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# How to find correlation between objects, in the dataframe as corr() gives correlation for only numeric data. 
# Creating a temporary data frame and can assign numeric values to the categorical data and compare them using corr() function. 
df_numeric = df
```

```python
for col in df_numeric:
    if df_numeric[col].dtype == 'object':
        df_numeric[col] = df_numeric[col].astype('category').cat.codes
```

```python
correl_matrix2 = df.corr()
correl_matrix2
```

```python
# the above matrix is also difficult to visualise so, we can use unstack to visualise better 
corr_pairs = correl_matrix2.unstack()
```
